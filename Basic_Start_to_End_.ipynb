```python
# questions : import tips dataset
import seaborn as sns
data =sns.load_dataset("tips")
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Count number of records per day
counts = data['day'].value_counts()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("Distribution of Records by Day")
plt.show()
```

```python
# Scatter plot
plt.scatter(data['total_bill'], data['tip'])
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.bar(data['total_bill'], data['tip'])
plt.title("bar Plot: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Sort the data for a proper line plot
data = data.sort_values("total_bill")

# Line plot
plt.plot(data['total_bill'], data['tip'])
plt.title("Line Plot: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```
